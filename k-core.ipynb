```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Definition of k-core

In graph theory, the k-core of a graph refers to a subgraph that is obtained by recursively removing nodes with degrees less than k. In other words, the k-core represents the maximal connected subgraph in which all nodes have a degree of at least k.

## Properties of the k-core


1. Higher k values yield a smaller and more interconnected k-core. As k increases, the number of nodes and edges in the k-core decreases, and the remaining nodes tend to be more densely connected.

2. The maximum value of k for a given graph is the degree of the node with the highest degree. This is because nodes with degrees less than k are removed during the k-core computation.

3. The k-core can be used to identify the most central and influential nodes in a graph. Nodes with higher degrees tend to be part of the higher-order k-cores, indicating their importance in maintaining the connectivity of the graph.

4. The k-core decomposition can help reveal the hierarchical organization of a graph. By examining different levels of k, one can identify nested structures or communities within the graph.

## Examples

```python
G = nx.karate_club_graph()
G.name = 'base'

H = nx.k_core(G, 2)
H.name = '2-core'

I = nx.k_core(G, 3)
I.name = '3-core'

J = nx.k_core(G, 4)
J.name = '4-core'
```

```python
print(nx.info(G), f', density={round(nx.density(G), 2)}', f'average_degree={round(G.number_of_edges()/G.number_of_nodes(), 2)}')
print(nx.info(H), f', density={round(nx.density(H), 2)}', f'average_degree={round(H.number_of_edges()/H.number_of_nodes(), 2)}')
print(nx.info(I), f', density={round(nx.density(I), 2)}', f'average_degree={round(I.number_of_edges()/I.number_of_nodes(), 2)}')
print(nx.info(J), f', density={round(nx.density(J), 2)}', f'average_degree={round(J.number_of_edges()/J.number_of_nodes(), 2)}')
pos = nx.kamada_kawai_layout(G)


fig, axis = plt.subplots(nrows=1, ncols=4, figsize=(12, 3), )
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.1, hspace=0)

nx.draw(G, pos=pos, node_size=150, with_labels=True, ax=axis[0])
nx.draw(H, pos=pos, node_size=150, with_labels=True, ax=axis[1])
nx.draw(I, pos=pos, node_size=150, with_labels=True, ax=axis[2])
nx.draw(J, pos=pos, node_size=150, with_labels=True, ax=axis[3])
```
