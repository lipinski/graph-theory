```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

# Kronocker Graph 

## Deterministic

```python
def generate_kronocker_graph_deterministic(init, n):
    A = np.matrix(init)

    for i in range(n):
        A = np.kron(A, init)
        
    return A
```

```python
kron_params = [
    [0, 1, 1],
    [1, 0, 1],
    [1, 1, 0]
]

A = generate_kronocker_graph_deterministic(kron_params, 2)
```

```python
G = nx.from_numpy_matrix(A)
```

```python
plt.matshow(A, cmap='Greys')
plt.show()
```

```python
fig = plt.subplots(figsize=(4, 4))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos, node_size=40)
```

## Stochastic

The solution is simple and the process is memory intensive.

```python
def generate_kronocker_graph_stochastic(init, n):
    A = init

    for i in range(n):
        A = np.kron(A, init)
        
    A = rand = np.random.binomial(1, A, A.shape) 
    
    return A
```

```python
kron_params = [
    [0.2, 0.5, 0.9],
    [0.6, 0.4, 0.6],
    [0.9, 0.8, 0.3]
]

A = generate_kronocker_graph_stochastic(kron_params, 2)
```

```python
G = nx.from_numpy_matrix(A)
```

```python
plt.matshow(A, cmap='Greys')
plt.show()
```

```python
fig = plt.subplots(figsize=(4, 4))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos, node_size=40)
```
